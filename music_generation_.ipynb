```python
# -*- coding: utf-8 -*-
"""CodeAlpha AI Internship - Task 3: Music Generation with AI - FIXED VERSION.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/your-notebook-link

# üéµ AI Music Generation with LSTM Neural Networks - FIXED VERSION

## CodeAlpha AI Internship - Task 3

This notebook implements a complete AI music generation system using LSTM neural networks.
"""

# Install required packages
!pip install music21 pretty_midi tensorflow midi2audio

import tensorflow as tf
import numpy as np
import music21
from music21 import *
import pretty_midi
import os
import glob
from collections import Counter
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from google.colab import files
import io

print("‚úÖ All packages installed and imported successfully!")
print(f"TensorFlow version: {tf.__version__}")

# Check if GPU is available
print(f"GPU Available: {tf.config.list_physical_devices('GPU')}")

"""## üì• Step 1: Download Better MIDI Dataset"""

# Download guaranteed-working MIDI files from a reliable source
print("üì• Downloading sample MIDI files...")

# Classical piano pieces that work well
!wget -q -O beethoven.mid "https://bitmidi.com/uploads/33966.mid"
!wget -q -O mozart.mid "https://bitmidi.com/uploads/81293.mid"
!wget -q -O bach.mid "https://bitmidi.com/uploads/75293.mid"

# Also try uploading your own files
print("üìÅ You can also upload your own MIDI files (optional):")
try:
    uploaded = files.upload()
    print(f"‚úÖ Uploaded {len(uploaded)} files")
except:
    print("‚ÑπÔ∏è No files uploaded, using sample files only")
    uploaded = {}

# List all MIDI files
midi_files = glob.glob("*.mid") + list(uploaded.keys())
print(f"üéπ Found {len(midi_files)} MIDI files: {[f for f in midi_files]}")

"""## üéº Step 2: Robust MIDI Parsing with Multiple Methods"""

def parse_midi_robust(midi_files):
    """Parse MIDI files using multiple methods to handle different formats"""
    notes = []
    successful_files = 0

    for file_path in midi_files:
        try:
            print(f"üìñ Attempting to parse: {file_path}")

            # Method 1: Try with music21 first
            try:
                midi = converter.parse(file_path)
                notes_to_parse = midi.flat.notes

                file_notes = []
                for element in notes_to_parse:
                    if isinstance(element, note.Note):
                        file_notes.append(str(element.pitch))
                    elif isinstance(element, chord.Chord):
                        file_notes.append('.'.join(str(n) for n in element.normalOrder))

                if file_notes:
                    notes.extend(file_notes)
                    successful_files += 1
                    print(f"   ‚úÖ Method 1: Extracted {len(file_notes)} notes")
                    continue

            except Exception as e:
                print(f"   ‚ùå Method 1 failed: {e}")

            # Method 2: Try with pretty_midi
            try:
                midi_data = pretty_midi.PrettyMIDI(file_path)
                file_notes = []

                for instrument in midi_data.instruments:
                    for note in instrument.notes:
                        # Convert pitch number to note name
                        note_name = pretty_midi.note_number_to_name(note.pitch)
                        file_notes.append(note_name)

                if file_notes:
                    notes.extend(file_notes)
                    successful_files += 1
                    print(f"   ‚úÖ Method 2: Extracted {len(file_notes)} notes")
                    continue

            except Exception as e:
                print(f"   ‚ùå Method 2 failed: {e}")

            # Method 3: Try parsing as stream directly
            try:
                midi_stream = converter.parse(file_path)
                file_notes = []

                for element in midi_stream.recurse():
                    if 'Note' in element.classes:
                        file_notes.append(str(element.pitch))
                    elif 'Chord' in element.classes:
                        file_notes.append('.'.join(str(n) for n in element.normalOrder))

                if file_notes:
                    notes.extend(file_notes)
                    successful_files += 1
                    print(f"   ‚úÖ Method 3: Extracted {len(file_notes)} notes")

            except Exception as e:
                print(f"   ‚ùå Method 3 failed: {e}")
                print(f"   üí° File {file_path} might be corrupted or in unsupported format")

        except Exception as e:
            print(f"‚ùå All methods failed for {file_path}: {e}")

    print(f"\nüìä Summary: Successfully parsed {successful_files}/{len(midi_files)} files")
    return notes

# Parse MIDI files with robust method
print("üéµ Extracting notes with robust parser...")
all_notes = parse_midi_robust(midi_files)

# If no notes extracted, use synthetic training data
if len(all_notes) == 0:
    print("‚ö†Ô∏è No notes extracted from MIDI files. Using synthetic training data...")
    # Create synthetic music data (scale patterns)
    scales = {
        'C Major': ['C', 'D', 'E', 'F', 'G', 'A', 'B'],
        'A Minor': ['A', 'B', 'C', 'D', 'E', 'F', 'G']
    }

    for scale_name, scale_notes in scales.items():
        # Create scale patterns
        for i in range(20):
            pattern = scale_notes * 3  # Repeat scale
            all_notes.extend(pattern)

        # Create arpeggio patterns
        for i in range(15):
            arpeggio = [scale_notes[0], scale_notes[2], scale_notes[4]] * 4
            all_notes.extend(arpeggio)

    print(f"‚úÖ Created {len(all_notes)} synthetic notes for training")

print(f"‚úÖ Total notes available: {len(all_notes)}")
print(f"üéπ Sample notes: {all_notes[:20]}")

# Analyze the music data
if all_notes:
    note_counts = Counter(all_notes)
    print(f"\nüìä Music Analysis:")
    print(f"Total notes: {len(all_notes)}")
    print(f"Unique notes/chords: {len(note_counts)}")
    print(f"Most common notes: {note_counts.most_common(10)}")

    # Plot note distribution
    plt.figure(figsize=(12, 6))
    top_notes = dict(note_counts.most_common(15))
    plt.bar(range(len(top_notes)), list(top_notes.values()))
    plt.xticks(range(len(top_notes)), list(top_notes.keys()), rotation=45)
    plt.title('Top 15 Most Frequent Notes/Chords')
    plt.tight_layout()
    plt.show()
else:
    print("‚ùå No notes available for training!")

"""## üîß Step 3: Prepare Data for LSTM Training"""

def prepare_sequences(notes, sequence_length=100):
    """Prepare input sequences and outputs for the LSTM model"""

    if len(notes) < sequence_length + 1:
        print(f"‚ö†Ô∏è Not enough notes. Have {len(notes)}, need at least {sequence_length + 1}")
        # Create repeating pattern to have enough data
        repeated_notes = notes * (sequence_length // len(notes) + 2)
        notes = repeated_notes[:sequence_length + 100]
        print(f"‚úÖ Extended to {len(notes)} notes")

    # Get all pitch names
    pitchnames = sorted(set(notes))

    # Create a dictionary to map pitches to integers
    note_to_int = dict((note, number) for number, note in enumerate(pitchnames))

    network_input = []
    network_output = []

    # Create input sequences and the corresponding outputs
    for i in range(0, len(notes) - sequence_length, 1):
        sequence_in = notes[i:i + sequence_length]
        sequence_out = notes[i + sequence_length]
        network_input.append([note_to_int[char] for char in sequence_in])
        network_output.append(note_to_int[sequence_out])

    n_patterns = len(network_input)

    if n_patterns == 0:
        raise ValueError("No training patterns generated. Check your data.")

    # Reshape the input into a format compatible with LSTM layers
    network_input = np.reshape(network_input, (n_patterns, sequence_length, 1))

    # Normalize input between 0 and 1
    network_input = network_input / float(len(pitchnames))

    # Convert output to categorical one-hot encoding
    network_output = to_categorical(network_output)

    return network_input, network_output, pitchnames, note_to_int

# Prepare training data
print("üîß Preparing training sequences...")
sequence_length = 50  # Reduced for smaller datasets
try:
    network_input, network_output, pitchnames, note_to_int = prepare_sequences(all_notes, sequence_length)

    print(f"‚úÖ Data preparation complete!")
    print(f"Input shape: {network_input.shape}")
    print(f"Output shape: {network_output.shape}")
    print(f"Vocabulary size: {len(pitchnames)}")
    print(f"Number of training sequences: {len(network_input)}")

except Exception as e:
    print(f"‚ùå Error in data preparation: {e}")
    print("üîÑ Creating fallback training data...")

    # Create simple fallback data
    simple_notes = ['C', 'D', 'E', 'F', 'G'] * 100
    network_input, network_output, pitchnames, note_to_int = prepare_sequences(simple_notes, 10)

"""## üß† Step 4: Build LSTM Neural Network"""

def create_music_model(input_shape, output_size, lstm_units=256, dropout_rate=0.3):
    """Create LSTM model for music generation"""

    model = Sequential()

    # First LSTM layer
    model.add(LSTM(
        lstm_units,
        input_shape=input_shape,
        return_sequences=True
    ))
    model.add(Dropout(dropout_rate))
    model.add(BatchNormalization())

    # Second LSTM layer
    model.add(LSTM(lstm_units))
    model.add(Dropout(dropout_rate))
    model.add(BatchNormalization())

    # Dense layers
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(dropout_rate))

    # Output layer
    model.add(Dense(output_size, activation='softmax'))

    # Compile model
    optimizer = Adam(learning_rate=0.001)
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

    return model

# Create the model
print("üß† Building LSTM neural network...")
try:
    model = create_music_model(
        input_shape=(network_input.shape[1], network_input.shape[2]),
        output_size=network_output.shape[1]
    )

    model.summary()
except Exception as e:
    print(f"‚ùå Error creating model: {e}")

"""## üèãÔ∏è Step 5: Train the Model"""

# Training parameters
epochs = 50  # Reduced for demonstration
batch_size = 32

print(f"üöÄ Starting training for {epochs} epochs...")

try:
    # Train the model
    history = model.fit(
        network_input,
        network_output,
        epochs=epochs,
        batch_size=batch_size,
        verbose=1,
        validation_split=0.2
    )

    # Plot training history
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend()

    plt.tight_layout()
    plt.show()

    print("‚úÖ Model training completed!")

    # Save the trained model
    model.save('music_generation_model.h5')
    print("üíæ Model saved as 'music_generation_model.h5'")

except Exception as e:
    print(f"‚ùå Training error: {e}")

"""## üéµ Step 6: Generate New Music"""

def generate_music(model, network_input, pitchnames, sequence_length=50, generation_length=200, temperature=1.0):
    """Generate new music using the trained model"""

    # Create reverse mapping from integers to notes
    int_to_note = dict((number, note) for number, note in enumerate(pitchnames))

    # Pick a random sequence from the input as a starting point
    start = np.random.randint(0, len(network_input)-1)
    pattern = network_input[start]
    prediction_output = []

    print("üéπ Generating music...")

    # Generate notes
    for note_index in range(generation_length):
        prediction_input = np.reshape(pattern, (1, len(pattern), 1))
        prediction_input = prediction_input / float(len(pitchnames))

        prediction = model.predict(prediction_input, verbose=0)[0]

        # Apply temperature for more creative/random results
        prediction = np.log(prediction) / temperature
        exp_preds = np.exp(prediction)
        prediction = exp_preds / np.sum(exp_preds)

        # Sample from the probability distribution
        index = np.random.choice(len(prediction), p=prediction)
        result = int_to_note[index]
        prediction_output.append(result)

        # Update the pattern
        pattern = np.append(pattern, index / float(len(pitchnames)))
        pattern = pattern[1:len(pattern)]

    print(f"‚úÖ Generated {len(prediction_output)} notes!")
    return prediction_output

# Generate music
print("üéµ Creating new musical composition...")
try:
    generated_notes = generate_music(
        model,
        network_input,
        pitchnames,
        sequence_length,
        generation_length=200,
        temperature=0.8
    )

    print(f"üéπ First 30 generated notes: {generated_notes[:30]}")
except Exception as e:
    print(f"‚ùå Music generation error: {e}")
    # Create simple fallback melody
    generated_notes = ['C', 'E', 'G', 'C', 'D', 'F', 'A', 'D', 'E', 'G', 'B', 'E'] * 10
    print("üîÑ Using fallback melody")

"""## üíæ Step 7: Convert to MIDI File"""

def create_midi_from_notes(prediction_output, filename="ai_generated_music.mid"):
    """Convert generated notes to a MIDI file"""

    offset = 0
    output_notes = []

    # Create note and chord objects
    for pattern in prediction_output:
        # Pattern is a chord (contains dots or numbers)
        if '.' in pattern:
            notes_in_chord = pattern.split('.')
            notes = []
            for current_note in notes_in_chord:
                try:
                    new_note = note.Note(int(current_note))
                    new_note.storedInstrument = instrument.Piano()
                    notes.append(new_note)
                except:
                    # If chord parsing fails, use middle C
                    new_note = note.Note(60)  # Middle C
                    new_note.storedInstrument = instrument.Piano()
                    notes.append(new_note)
            new_chord = chord.Chord(notes)
            new_chord.offset = offset
            output_notes.append(new_chord)
        else:
            # Pattern is a note
            try:
                new_note = note.Note(pattern)
                new_note.offset = offset
                new_note.storedInstrument = instrument.Piano()
                output_notes.append(new_note)
            except:
                # If note parsing fails, use middle C
                new_note = note.Note(60)  # Middle C
                new_note.offset = offset
                new_note.storedInstrument = instrument.Piano()
                output_notes.append(new_note)

        # Increase offset
        offset += 0.5

    # Create MIDI stream
    midi_stream = stream.Stream(output_notes)

    # Save MIDI file
    midi_stream.write('midi', fp=filename)

    return filename

# Create MIDI file
print("üíæ Creating MIDI file...")
try:
    midi_filename = create_midi_from_notes(generated_notes)
    print(f"‚úÖ MIDI file created: {midi_filename}")

    # Download the generated music
    files.download(midi_filename)
    print("üì• MIDI file downloaded to your computer!")
except Exception as e:
    print(f"‚ùå MIDI creation error: {e}")

"""## üéß Step 8: Play the Generated Music"""

def play_midi_in_colab(midi_file):
    """Convert MIDI to audio and play in Colab"""
    try:
        # Convert MIDI to WAV using pretty_midi
        midi_data = pretty_midi.PrettyMIDI(midi_file)
        audio_data = midi_data.synthesize()

        # Play audio
        from IPython.display import Audio
        return Audio(audio_data, rate=44100)
    except Exception as e:
        print(f"‚ùå Could not generate audio: {e}")
        return None

# Play the generated music
print("üéß Generating audio preview...")
try:
    audio_preview = play_midi_in_colab(midi_filename)
    if audio_preview:
        print("üîä Playing generated music...")
        display(audio_preview)
    else:
        print("üí° Audio generation failed. Download the MIDI file to play locally.")
except Exception as e:
    print(f"‚ùå Audio playback error: {e}")

"""## üìä Step 9: Analyze Generated Music"""

# Analyze the generated music
if 'generated_notes' in locals():
    generated_note_counts = Counter(generated_notes)

    print("üìä Generated Music Analysis:")
    print(f"Total notes generated: {len(generated_notes)}")
    print(f"Unique notes/chords: {len(generated_note_counts)}")
    print(f"Most common generated notes: {generated_note_counts.most_common(10)}")

    # Compare original vs generated music
    plt.figure(figsize=(15, 5))

    # Original music
    plt.subplot(1, 2, 1)
    top_original = dict(Counter(all_notes).most_common(8))
    plt.bar(range(len(top_original)), list(top_original.values()))
    plt.xticks(range(len(top_original)), list(top_original.keys()), rotation=45)
    plt.title('Original Music - Top Notes')

    # Generated music
    plt.subplot(1, 2, 2)
    top_generated = dict(generated_note_counts.most_common(8))
    plt.bar(range(len(top_generated)), list(top_generated.values()))
    plt.xticks(range(len(top_generated)), list(top_generated.keys()), rotation=45)
    plt.title('Generated Music - Top Notes')

    plt.tight_layout()
    plt.show()

"""## üéØ Step 10: Generate Multiple Variations"""

print("üéπ Generating multiple music variations...")

# Generate different pieces with different temperatures
temperatures = [0.5, 1.0, 1.5]  # Low, medium, high creativity

for i, temp in enumerate(temperatures):
    print(f"\nüéµ Generating variation {i+1} with temperature {temp}...")

    try:
        variation_notes = generate_music(
            model,
            network_input,
            pitchnames,
            sequence_length,
            generation_length=150,
            temperature=temp
        )

        # Create MIDI file
        variation_filename = f"ai_music_variation_{i+1}.mid"
        create_midi_from_notes(variation_notes, variation_filename)

        # Download
        files.download(variation_filename)
        print(f"‚úÖ Downloaded: {variation_filename}")
    except Exception as e:
        print(f"‚ùå Error generating variation {i+1}: {e}")

"""## üìù Project Summary

### ‚úÖ Task 3 Completed Successfully!

**What we accomplished:**
1. ‚úÖ **MIDI Data Collection** - Downloaded and parsed piano music
2. ‚úÖ **Robust Data Preprocessing** - Multiple parsing methods with fallbacks
3. ‚úÖ **LSTM Model Building** - Created neural network for music generation
4. ‚úÖ **Model Training** - Trained on musical patterns
5. ‚úÖ **Music Generation** - Generated original compositions
6. ‚úÖ **MIDI Export** - Created downloadable MIDI files
7. ‚úÖ **Audio Playback** - Listen to generated music
8. ‚úÖ **Multiple Variations** - Different creativity levels

### üéµ Generated Outputs:
- `ai_generated_music.mid` - Main composition
- `ai_music_variation_1.mid` - Conservative
- `ai_music_variation_2.mid` - Balanced
- `ai_music_variation_3.mid` - Creative

### üîß Technical Features:
- **Robust MIDI parsing** with multiple fallback methods
- **Synthetic data generation** if MIDI files fail
- **Error handling** at every step
- **Visual analysis** of music patterns
- **Temperature control** for creativity

---

**CodeAlpha AI Internship - Task 3: Music Generation with AI - COMPLETED üöÄ**
"""

# Final download of all project files
print("üì¶ Creating project archive...")
try:
    !zip -r music_generation_project.zip *.mid *.h5 2>/dev/null || echo "No files to archive"
    files.download('music_generation_project.zip')
    print("‚úÖ Project files archived and downloaded!")
except:
    print("‚ÑπÔ∏è Could not create archive, but individual files are available")

print("üéâ Project completed successfully!")
print("üìã Ready for submission to CodeAlpha!")
```
